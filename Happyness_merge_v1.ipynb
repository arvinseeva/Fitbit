```python
#This Jupyter Notebook is a result of our collaboration
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#read in data
df1_path = "Resources/2015.csv"
df2_path = "Resources/2016.csv"
df3_path = "Resources/2017.csv"
df4_path = "Resources/2018.csv"
df5_path = "Resources/2019.csv"
df6_path = "Resources/2020.csv"

df1 = pd.read_csv(df1_path)
df2 = pd.read_csv(df2_path)
df3 = pd.read_csv(df3_path)
df4 = pd.read_csv(df4_path)
df5 = pd.read_csv(df5_path)
df6 = pd.read_csv(df6_path)

df6.info()
```

```python
#combine 2020 & 2019 dataframes together
data = pd.merge(df6,df5, how='left', left_on=['Country name'], right_on=['Country or region'])
data.head()
```

```python
#rename columns of 2020 & 2019
data = data.rename(columns={'Social support_2020':"Social support 2020", "Healthy life expectancy_2020":"Healthy life expectancy 2020",
                            'Freedom to make life choices_x':'Freedom to make life choices 2020',
                           'Social support_y':'Social support 2019', 'Healthy life expectancy_y': 'Healthy life expectancy 2019',
                           'Freedom to make life choices_y':'Freedom to make life choices 2019', 'Generosity_y':'Generosity 2019',
                           'Perceptions of corruption_y':'Perceptions of corruption 2019'})
data.head()
```

```python
#adding 2018 dataset to current dataset (data)
data = pd.merge(data, df4, how='left', left_on=['Country name'], right_on=['Country or region'])
data.head()
```

```python
#rename columns of 2018
data = data.rename(columns={'Overall rank_y':'Overall rank 2018', 'Score_y':'Score 2019', 'GDP per capita_y': 'GDP per capital 2018' })
data.head()
```

```python
# Rename columns 
# data.rename(columns={0:'Countryxxxx',1:'Yearxxxx'},inplace=True)
# data

data.columns = data.columns.str.replace('_x',' 2020')
data.info()
```
